Fragmentos de Morgan

```python
# Fragmento de morgan 273
from rdkit.Chem import Draw
from IPython.display import display_svg
from rdkit import Chem
from rdkit.Chem import AllChem
import csv

csv_file_path = "./../Dados/QM-symex-DB.csv"

with open(csv_file_path , "r") as csvfile:
    csv_reader = csv.reader(csvfile)
    next(csv_reader, None)

    rows = list(csv_reader)
    for index, row in enumerate(rows[48000:50000]):
        try:
            mol = row[0]
            mol = Chem.MolFromSmiles(mol)
            
            # Gerando a fingerprint de Morgan com raio 2 e 512 bits
            fpgen = AllChem.GetMorganGenerator(radius=2)
            ao = AllChem.AdditionalOutput()
            ao.CollectBitInfoMap()
            fp = fpgen.GetFingerprint(mol, additionalOutput=ao)
            bi = ao.GetBitInfoMap()
            for i in range(1):
                
                
                    bit_position = i + 273 # Escolha o bit especfico que deseja imprimir
                    print(bi[bit_position])
                    mfp_svg = Draw.DrawMorganBit(mol, bit_position, bi, useSVG=True)
                    display_svg(mfp_svg)
        except:
            KeyError
```
