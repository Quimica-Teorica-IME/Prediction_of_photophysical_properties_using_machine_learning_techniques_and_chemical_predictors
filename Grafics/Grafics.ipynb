```python
import matplotlib.pyplot as plt
import csv
import numpy as np

csv_file_path = "./../Dados/Exp-DB.csv"

pa_list = []
Fl_list = []


with open(csv_file_path , "r") as csvfile:
    csv_reader = csv.reader(csvfile)
    next(csv_reader, None)
    
    rows = list(csv_reader)
    for row in rows:
        Fl = float(row[6])
        Fl_list.append(Fl)

Fl_list = np.array(Fl_list)

# Dados de exemplo
valores = [0.5, 0.6, 0.3, 0.7, 0.1, 0.5, 0.2, 0.8, 0.4]

# Defina os intervalos como valores decimais de 0.1
intervalos = [round(i * 0.1, 1) for i in range(12)]

# Inicialize a contagem de valores em cada intervalo
contagem_valores = [0] * len(intervalos)
contagem = 0
# Conte os valores em cada intervalo
for valor in Fl_list:
    if valor > 0.1:
        contagem += 1
    for i, intervalo in enumerate(intervalos):
        if i < len(intervalos) - 1 and intervalo <= valor < intervalos[i + 1]:
            contagem_valores[i] += 1
print(contagem)
# Crie um gráfico de colunas
plt.bar(intervalos, contagem_valores, width=0.1, edgecolor='black', color='orange')

# Adicione rótulos ao gráfico
plt.xlabel('Medida rendimento quântico')
plt.ylabel('Contagem de Valores')
plt.title('Gráfico de Comprimento de onda')

# Exiba o gráfico
plt.show()

```
